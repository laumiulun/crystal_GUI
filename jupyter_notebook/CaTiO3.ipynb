```python
#GlowScript 2.8 VPython

from vpython import *


scene5=canvas(background = color.black, width=600, length=600, center=vector(-1,-1,-1))
#-------------------------------------


#Ca2+: 1.14
#Ti4+: 0.75
#O2-:1.21
#a: 3.8967
#bond length: Ca-O 2.702 O-Ti:1.949, Ca-Ca:2.622

L = 1
bond_ab = 3.79
ra_int = 1.14
rb_int = 0.75
ro_int = 1.21

bond_default_text = str(bond_ab)
ra_default_text = str(ra_int)
rb_default_text = str(rb_int)
ro_default_text = str(ro_int)


scale_factor = 0.5
scale_default_text = str(scale_factor)
#--------------------------------------
ra = ra_int/bond_ab*scale_factor
rb = rb_int/bond_ab*scale_factor
ro = ro_int/bond_ab*scale_factor

ra_item = []
rb_item = []
ro_item = []
cylinder_item = []
total_item = []

running = False

opatom = 1
opbar = 1


## need to change the input parameters
def build(L,ra,rb,ra_item,rb_item,ro_item,cylinder_item,total_item):
    L1 = L
    LB = L-1
    LN1 = L-1
    #global total_item
    for i in range(-L, L1):
        for j in range(-L, L1):
            for k in range(-L, L1):
                # plot A: Green
                for na in range(-1,0):
                    v0=vector(i+1/2*na,j+1/2*na,k+1/2*na)
                    b=sphere(pos = v0, radius = ra, color = color.green, opacity=opatom) 
                    ra_item.append(b)
                    if i< LN1:
                        b=cylinder(pos = v0, axis=vector(1,0,0),radius = .01, color = color.white, opacity=opbar)
                        cylinder_item.append(b)
                    if j< LN1:
                        b=cylinder(pos = v0, axis=vector(0,1,0),radius = .01, color = color.white, opacity=opbar)
                        cylinder_item.append(b)
                    if k< LN1:
                        b=cylinder(pos = v0, axis=vector(0,0,1),radius = .01, color = color.white, opacity=opbar)
                        cylinder_item.append(b)
                # plot B: Blue
                if(i<LB and j<LB and k<LB):
                    b=sphere(pos = vector(i,j,k), radius = rb, color = vector(0.0,0.7,1), opacity=opatom)
                    rb_item.append(b)
                    # plot O: red
                    v=[vector(i-1/2,j,k), vector(i+1/2,j,k), vector(i,j-1/2,k), vector(i,j+1/2,k), \
                       vector(i,j,k-1/2), vector(i,j,k+1/2)]
                    for x in v:
                        b=sphere(pos = x, radius = ro, color = color.red, opacity=opatom)
                        ro_item.append(b)
                    for m in range(2,6):
                        b=cylinder(pos = v[0], axis=v[m]-v[0],radius = .01, color = color.white, opacity=opbar)
                        cylinder_item.append(b)
                        b=cylinder(pos = v[1], axis=v[m]-v[1],radius = .01, color = color.white, opacity=opbar)
                        cylinder_item.append(b)
                    for m in range(4,6):
                        b=cylinder(pos = v[2], axis=v[m]-v[2],radius = .01, color = color.white, opacity=opbar)
                        cylinder_item.append(b)
                        b=cylinder(pos = v[3], axis=v[m]-v[3],radius = .01, color = color.white, opacity=opbar)
                        cylinder_item.append(b)
                    
    total_item = ra_item + rb_item + ro_item + cylinder_item
    #print(len(total_item),len(ra_item),len(rb_item))
    return ra_item,rb_item,ro_item,cylinder_item,total_item

def build_button(s):
    global ra_item
    global rb_item
    global ro_item
    global cylinder_item
    global total_item
    global running
    running = not running
    if running: s.text = "Pause"
    else: s.text = "Run"
    
    #ra_item,rb_item,cylinder_item,total_item = build(L,ra,rb,ra_item,rb_item,cylinder_item,total_item)
# clear item
def clear():
    for i in ra_item:
        i.visible = False
        del i
    for i in rb_item:
        i.visible = False
        del i 
    for i in ro_item:
        i.visible = False
        del i 
    for i in cylinder_item:
        i.visible = False
        del i

# resetting button, rebuild based on inital parameters
def reset():
    global ra_item
    global rb_item
    global ro_item
    global cylinder_item
    global total_item
    clear()
    ra_item,rb_item,ro_item,cylinder_item,total_item = build(L,ra,rb,ra_item,rb_item,ro_item,cylinder_item,total_item)

# Function for setting bond_ab
def set_bondab(s):
    bond_ab = s.number
    ra = ra_int/bond_ab*scale_factor
    rb = rb_int/bond_ab*scale_factor
    ro = ro_int/bond_ab*scale_factor
    for x in ra_item:
        x.radius = ra
    for y in rb_item:
        y.radius = rb
    for z in ro_item:
        z.radius = ro
    return bond_ab

# Function for setting r_a
def set_ra(s):
    ra_int = s.number 
    ra = ra_int/bond_ab*scale_factor
    for x in ra_item:
        x.radius = ra
    return ra
# Function for setting r_b
def set_rb(s):
    rb_int = s.number
    rb = rb_int/bond_ab*scale_factor
    for y in rb_item:
        y.radius = rb
    return rb

# Function for setting r_o
def set_ro(s):
    ro_int = s.number
    ro = ro_int/bond_ab*scale_factor
    for y in ro_item:
        y.radius = ro
    return ro

def set_scale_factor(s):
    scale_factor = s.number
    ra = ra_int/bond_ab*1*scale_factor
    rb = rb_int/bond_ab*1*scale_factor
    ro = ro_int/bond_ab*1*scale_factor
    for x in ra_item:
        x.radius = ra
    for y in rb_item:
        y.radius = rb
    for z in ro_item:
        z.radius = ro
    return scale_factor,ra,rb,ro

# Setting distance L
def set_L(s):
    global ra_item
    global rb_item
    global ro_item
    global cylinder_item
    global total_item
    L = s.number
    clear()
    ra_item,rb_item,ro_item,cylinder_item,total_item = build(L,ra,rb,ra_item,rb_item,ro_item,cylinder_item,total_item)
#     #total_item = ra_item + rb_item + cylinder_item
#     return ra_item,rb_item,cylinder_item#,total_item

# inital build 
ra_item,rb_item,ro_item,cylinder_item,total_item = build(L,ra,rb,ra_item,rb_item,ro_item,cylinder_item,total_item)
    #print(len(total_item),len(ra_item),len(rb_item))
    
# Heading explanation
#############################################################################################
wtext(text='<h2>Perovskite Example: Calcium Titanate (<b> CaTiO<sub>3</sub> </b>)</h2><hr>',pos= scene5.title_anchor)
wtext(text='''
A <b>Perovskite</b> is a colourless mineral of Calcium Titanate with the chemical formula <b> CaTiO<sub>3</sub> </b>. 
It is discovered by German mineralogist Gustave Rose who named it after a Russian mineralogist
Lev Alexeievitch Perovskite. Later, perovskites are referred to a family of materials with the 
same crystal structure as <b> CaTiO<sub>3</sub> </b>. So the general chemical formula for this category 
of perovskite compounds is <b> ABX<sub>3</sub> </b>, where A is a metal cation (positively charged ion), 
B is a cation and X is an anion (negatively charged ion).  If X is oxygen, the sub-category can also be 
represented as <b> ABO<sub>3</sub> </b>.

<b> CaTiO<sub>3</sub> </b> consists of three types of ions: positive Calcium ions (symbol: Ca+), Titanium 
ions (symbol: Ti+) and negative Oxygen ions (symbol: O-) which alternate with each other in each of the 
three dimensions. In a unit cell, Ca+ occupies the corners,  Ti+ occupies the center of the lattice and 
O- are at the center of faces. 

Similarly, <b> SrTiO<sub>3</sub> </b> (Strontium Titanate) consists of three types of ions: positive 
Strontium ions (symbol: Sr+), Titanium ions (symbol: Ti+) and negative Oxygen ions (symbol: O-) in the 
same way. It is one of the few materials with ideal perovskite structure (no slight distortion). 

Perovskites are of great importance in material science and can be widely applied to photovoltaics, laser, 
semiconductors and so on. 

<b>Perovskite solar cells</b>: Solar cells with perovskite structures have shown remarkable progress in 
improving energy conversion efficiency, from reports of about 3% in 2006 to over 24% in 2019 and thus are 
thought to be a frontier energy solution.  Department of Energy is of great interest in research to increase 
the efficiency and lifetime of hybrid organic-inorganic perovskite solar cells. Targeted materials include 
methylammonium lead halide, i.e., <b> CH<sub>3</sub>NH<sub>3</sub>PbX<sub>3</sub> </b>, where X = I, Br or Cl.

There are two GUI files for Ideal Perovskite Structure solids. One for Calcium Titanate (<b> CaTiO<sub>3</sub> </b>), and 
one for Strontium Titanate (<b> SrTiO<sub>3</sub> </b>)

''',pos= scene5.title_anchor)

wtext(text='''<h3>Control</h3>\
<ul>\
<li><b>Rotate</b>: Right Button drag or Ctrl-drag</li>\
<li><b>Zoom</b>: Drag with middle button or use scroll wheel</li>\
</ul>''',pos= scene5.title_anchor)

wtext(text='''<h3>Modifying Structure</h3>\
<ul>\
<li><b>L</b>: Number of Unit Cells(Integar Only)</li>\
<li><b>bond</b>: distance between two closest Ca+ ions </li>\
<li><b><font color ="Green">r_a<font color ="black"></b>: ionic radius of Ca+ [<font color = "Green">Green<font color ="black">] </li>\
<li><b><font color ="Blue">r_b<font color ="black"></b>: ionic radius of Ti+ [<font color ="Blue">Blue<font color ="black">] </li>\
<li><b><font color ="Red">r_o<font color ="black"></b>: ionic radius of O- [<font color ="Red">Red<font color ="black">] </li>\
</ul>''',pos= scene5.title_anchor)

wtext(text='<hr>',pos = scene5.title_anchor)
#############################################################################################
# Button
b1 = button(text = "Run", pos = scene5.title_anchor,bind = build_button)
b2 = button(text = "Reset", pos = scene5.title_anchor,bind = reset)

it0= wtext(text='<h5>Parameters:</h5>Adjust the parameters using the input box(press Enter key after modifying parameters every time)\n')


# Parameters input box
it1 = wtext(text='L: ')
ib1 = winput(type = "numeric",pos=scene5.caption_anchor,bind=set_L,text=1)
scene5.append_to_caption("\n")

it2 = wtext(text='bond: ')
ib2 = winput(type = "numeric",pos=scene5.caption_anchor,bind=set_bondab, text = bond_default_text)
wtext(text = "\\(   \\unicode{x212B}   \\)")
scene5.append_to_caption("\n")

it3 = wtext(text = 'ra: ')
ib3 = winput(type = "numeric",pos=scene5.caption_anchor,bind=set_ra,text = ra_default_text)
wtext(text = "\\(   \\unicode{x212B}   \\)")
scene5.append_to_caption("\n")

it4 = wtext(text = 'rb: ')
ib4 = winput(type = "numeric",pos=scene5.caption_anchor,bind=set_rb,text=rb_default_text)
wtext(text = "\\(   \\unicode{x212B}   \\)")
scene5.append_to_caption("\n")

it5 = wtext(text = 'ro: ')
ib5 = winput(type = "numeric",pos=scene5.caption_anchor,bind=set_ro,text=ro_default_text)
wtext(text = "\\(   \\unicode{x212B}   \\)")
scene5.append_to_caption("\n")


# colorline atom
it6 = wtext(text = 'scale: ')
ib6 = winput(type = "numeric",pos=scene5.caption_anchor,bind=set_scale_factor,text=scale_default_text)
wtext(text = "( Must be less than or equal to 1. A value less than 1 is used for an exploded view. )")



#wo6 = wtext(text = rb)
scene5.append_to_caption("\n\n")


MathJax.Hub.Queue(["Typeset",MathJax.Hub])

# NEED TO FIX ROTATION, since the item is not returning
while True:
    rate(100)
    if running:
        for b in total_item:
            b.rotate(angle=.01, axis=vector(0,1,0), origin=vector(0,0,0))



```
